```python
plt.hist(inclusive_energy, bins=50, alpha=0.5, label="Inclusive", density=True)
plt.hist(semi_energy, bins=50, alpha=0.5, label="Semi-Inclusive", density=True)

plt.xlabel("Energy")
plt.ylabel("Normalized Count")
plt.title("Energy Distribution (Normalized)")
plt.legend()
plt.yscale("log")
plt.xscale("log")
plt.show()
```

```python
bins = np.logspace(np.log10(min(energy)), np.log10(max(energy)), 50)
plt.hist(inclusive_energy, bins=bins, alpha=0.5, label="Inclusive", density=True)
plt.hist(semi_energy, bins=bins, alpha=0.5, label="Semi-Inclusive", density=True)

plt.xscale("log")
plt.yscale("log")
```
