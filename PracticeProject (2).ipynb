```python
def raw_ratio(df_all, df_species):
    all_counts, _ = np.histogram(df_all["log10E"], bins=bins)
    species_counts, _ = np.histogram(df_species["log10E"], bins=bins)

    ratio = np.zeros_like(all_counts, dtype=float)
    mask = all_counts > 0
    ratio[mask] = species_counts[mask] / all_counts[mask]
    return ratio
```

```python
def weighted_ratio(df_all, df_species):
    all_w, _ = np.histogram(df_all["log10E"], bins=bins, weights=df_all["weight_total"])
    species_w, _ = np.histogram(df_species["log10E"], bins=bins, weights=df_species["weight_total"])

    ratio = np.zeros_like(all_w, dtype=float)
    mask = all_w > 0
    ratio[mask] = species_w[mask] / all_w[mask]
    return ratio
```

```python
def meson_fraction(df_all_mesons, df_species):
    all_counts, _ = np.histogram(df_all_mesons["log10E"], bins=bins)
    species_counts, _ = np.histogram(df_species["log10E"], bins=bins)

    ratio = np.zeros_like(all_counts, dtype=float)
    mask = all_counts > 0
    ratio[mask] = species_counts[mask] / all_counts[mask]
    return ratio
```

```python
def meson_fraction_weighted(df_all_mesons, df_species):
    all_w, _ = np.histogram(df_all_mesons["log10E"], bins=bins, weights=df_all_mesons["weight_total"])
    species_w, _ = np.histogram(df_species["log10E"], bins=bins, weights=df_species["weight_total"])

    ratio = np.zeros_like(all_w, dtype=float)
    mask = all_w > 0
    ratio[mask] = species_w[mask] / all_w[mask]
    return ratio
```

```python
plt.figure()

for pdg in meson_list:
    df_s = df_charm_mesons[df_charm_mesons["abs_charm"] == pdg]
    r = meson_fraction_weighted(df_meson_total, df_s)
    plt.plot(bin_centers, r, marker="o", label=pdg_labels[pdg])

plt.xlabel("log10(E)")
plt.ylabel("Fraction within Mesons")
plt.title("Meson Composition vs Energy")
plt.legend()
plt.show()
```
